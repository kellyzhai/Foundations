import datetime

import datetime 
start = datetime.datetime(2012, 7, 31) 
end = datetime.datetime(2017, 6, 30) 

```python
aapl = web.DataReader('AAPL', 'google', start, end)
aapl.head()
```

```python
aapl.to_csv('aapl.csv')
aapl['Close'].plot(figsize=(10, 8)) 
plt.show()
```

```python
bbry = web.DataReader('BBRY', 'google', start, end)
bbry
bbry.to_csv('bbry.csv')
bbry['Close'].plot(figsize=(10, 8)) 
plt.show()
```

```python
lulu = web.DataReader('lulu', 'google', start, end)
lulu
lulu.to_csv('lulu.csv')
lulu['Close'].plot(figsize=(10, 8)) 
plt.show()
```

```python
amzn = web.DataReader('AMZN', 'google', start, end)
amzn
amzn.to_csv('amzn.csv')
amzn['Close'].plot(figsize=(10, 8)) 
plt.show()
```

```python
import pandas as pd
stocks = pd.DataFrame({"AAPL": aapl['Close'],
                      "BBRY": bbry['Close'],
                      "LULU": lulu['Close'],
                      "AMZN": amzn['Close']})
stocks.head()

all_monthlys = pd.date_range(start, end, freq='M')

stocks = stocks.reindex(all_monthlys)
stocks = stocks.fillna(method='ffill')
stocks.head(10)
```

```python
stocks.plot()
plt.show()
```

```python
#stock_return = stocks.apply(lambda x: x / x[0])
stocks_return = (stocks- stocks.shift(1))/stocks
stocks_return
```

```python
stocks_return.plot(grid = True).axhline(y = 1, color = "black", lw = 2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
from statsmodels.graphics import tsaplots

def label(ax, string):
   ax.annotate(string, (1, 1), xytext=(-8, -8), ha='right', va='top',
              size=14, xycoords='axes fraction', textcoords='offset points')

fig, axes = plt.subplots(nrows=4, figsize=(8, 12))
fig.tight_layout()


axes[0].acorr(stocks['AAPL'],maxlags=12)
label(axes[0], 'AAPL Autocorrelation')


axes[1].acorr(stocks['BBRY'],maxlags=12)
label(axes[1], 'BBRY Autocorrelation')

axes[2].acorr(stocks['LULU'],maxlags=12)
label(axes[2], 'LULU Autocorrelation')

axes[3].acorr(stocks['AMZN'],maxlags=12)
label(axes[3], 'AMZN Autocorrelation')

# Remove some of the titles and labels that were automatically added
for ax in axes.flat:
   ax.set(title='', xlabel='')

plt.show()
```

```python

from pandas import Series
from matplotlib import pyplot
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
plot_acf(stocks['AAPL'],label='AAPL ACF')
plot_pacf(stocks['AAPL'],label='AAPL ACF')
pyplot.show()

```

```python
from scipy import  stats
import statsmodels.api as sm  

m = 10
acf,q,p = sm.tsa.acf(stocks['AAPL'],nlags=m,qstat=True) 
out = np.c_[range(1,11), acf[1:], q, p]
output=pd.DataFrame(out, columns=['lag', "AC", "Q", "P-value"])
output = output.set_index('lag')
output
```
