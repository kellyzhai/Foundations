```python
d = data_str.split('\n')
d = [ i.split('\t') for i in d ]
```

```python
d
```

```python
for i in range( len( d ) ):
    for j in range( len( d[0] ) ):
        try:
            d[i][j] = float( d[i][j] )
        except:
            pass
```

```python
df = DataFrame( d[1:], columns=d[0])
```

```python
df
```

```python
#df['Eins'] = np.ones(( len(df), ))
Y = df.E[:-1]
X = df[['P','T']][:-1]
X = sm.add_constant(X)
result = sm.OLS( Y, X ).fit()
result.summary()
```
